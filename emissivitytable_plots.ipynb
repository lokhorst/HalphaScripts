```python
# log10 (E/nH^2) = log10(E) - log10(nH^2)
# log10 (0.2 * E/nH^2) = log10(0.2) + log10(E) - log10(nH^2)

SMALL_SIZE = 16
MEDIUM_SIZE = 16
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

reds = plt.get_cmap("Reds")
blues = plt.get_cmap("Blues")

def emisplot(ax,density_index = 40,legend=False,frac_of_solar = 1.0):
    'plot the H lines'
    indices = [1,9]
    linestyles = ['-','--']
    colours = [reds(1.0),reds(1.0)]
    for index,colour,linestyle in zip(indices,colours,linestyles):
        emtable = H_lines[:,:,index]
        ax.plot(H_logTK,emtable[:,density_index],label=H_wavelen[index],color=colour,linestyle=linestyle,linewidth=2)

    'plot the O lines'
    indices = [22,21,1,2,0]
    linestyles = ['-','--','-','--',':']
    colours = [blues(0.5),blues(0.5),blues(1.0),blues(1.0),blues(1.0)]
    for index,colour,linestyle in zip(indices,colours,linestyles):
        emtable = O_lines[:,:,index]
        label=O_wavelen[index]
        if 'O  3' in label:
            label = label.replace('O  3','[O  3]')
        ax.plot(O_logTK,emtable[:,density_index]+np.log10(frac_of_solar),label=label,color=colour,linestyle=linestyle,linewidth=2)

    'add a legend if set to True'
    if legend:
        ax.legend(fancybox=True,loc=1,fontsize=14)
        
def threeplotsatmet(ax1,ax2,ax3,frac_of_solar = 1.0,legend=True):
    indices = [22,21,1,2,0]; 
    linestyles = ['-','--','-','--',':']; 
    colours = [blues(0.5),blues(0.5),blues(1.0),blues(1.0),blues(1.0)]

    ### log10 n_H = 0
    emisplot(ax1,density_index = 40,frac_of_solar=frac_of_solar)
    
    ### log10 n_H = -3 
    emisplot(ax2,density_index = 25,frac_of_solar=frac_of_solar)

    ### log10 n_H = -6
    emisplot(ax3,density_index = 10,frac_of_solar=frac_of_solar,legend=legend)
        
    for ax,dens in zip([ax1,ax2,ax3],[0.0,-3.0,-6.0]):
        ax.set_xlim(3,8)
        ax.set_ylim(-30,-20)
        ax.set_xlabel(r'Log$_{10}$ T (K)')
        ax.set_ylabel(r'Log$_{10}$ $\epsilon / n_H^2$ ($erg$ $s^{-1}$ $cm^3$)')
        
        'add the density to the plot'
        label = r'log$_{10}$ n$_H$ = %s'%dens
        ax.text(0.05,0.94,label,transform=ax.transAxes,color='black',fontsize=SMALL_SIZE)
        
        'add the abundance to the plot'
        label = '[O/H] = %s'%np.log10(frac_of_solar)
        ax.text(0.05,0.87,label,transform=ax.transAxes,color='black',fontsize=SMALL_SIZE)
```
