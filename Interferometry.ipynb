```python
x = np.arange(np.pi/4.,3./4. *np.pi,0.01)
a = np.cos(x)
b = np.cos(x+0.1)
c = np.cos(x+1.)
d = np.cos(x+2.)
plt.plot(x,a)
#plt.plot(x,b)
#plt.plot(x,a*b)
plt.plot(x,np.cos(a))
plt.plot(x,np.cos(20.*a))
```
