## plot_SBhist

Compare the SB histograms of different box sizes of the EAGLE simulation.  Want to make sure that we are consistent between the different projection runs.

Framework taken from Halpha_SBhist.py.

```python
import numpy as np
import eagle_constants_and_units as c
import cosmo_utils as csu
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import mpl_toolkits.axes_grid1 as axgrid
from astropy import constants as const
from astropy import units as u

import get_halpha_SB

%matplotlib inline
```

```python
factor = 10
sl = [slice(None,None,None), slice(None,None,None)]
```

### Simulation snapnum 28 (z = 0), xy box size: 100Mpc, z slice width: 5Mpc,
Projections run by Nastasha.

```python
fileloc = '/Volumes/Cerulean/EAGLE/EagleProjections/EmissionMaps/FromNastasha/'
files_SF_28 = [fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen12.5__fromSFR.npz',
               fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen17.5__fromSFR.npz',
               fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen2.5__fromSFR.npz',
               fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen7.5__fromSFR.npz']
files_noSF_28 = [fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen12.5_noSFR.npz',
                 fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen17.5_noSFR.npz',
                 fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen2.5_noSFR.npz',
                 fileloc+'emission_halpha_L0100N1504_28_test2_SmAb_C2Sm_32000pix_5.000000slice_zcen7.5_noSFR.npz']
```

```python
print('Reading: '+files_noSF_28[0]+'...')
data2 = (np.load(files_noSF_28[0])['arr_0'])[sl]
data2 = get_halpha_SB.imreduce(data2, factor, log=True, method = 'average')
print('Reading" "'+files_SF_28[0]+'...')
data22 = (np.load(files_SF_28[0])['arr_0'])[sl]
data22 = get_halpha_SB.imreduce(data22, factor, log=True, method = 'average')
print('5 Mpc slice...')
data_28 = np.log10(10**data2+10**data22)
print('delete %s...'%(files_noSF_28[0]))
del data22
```

### Simulation snapnum 28 (z = 0), xy box size: 25Mpc, z slice width: 5Mpc,
Projections run by Deborah on niagara.

```python
fileloc_25Mpc = '/Volumes/Cerulean/EAGLE/EagleProjections/EmissionMaps/niagara_20Jan2019/halpha_25Mpc_376/'
files_SF_25Mpc = [fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen12.5__fromSFR.npz',
                  fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen17.5__fromSFR.npz',
                  fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen2.5__fromSFR.npz',
                  fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen7.5__fromSFR.npz',
                  fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_total__fromSFR.npz']
files_noSF_25Mpc = [fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen12.5_noSFR.npz',
                    fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen17.5_noSFR.npz',
                    fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen2.5_noSFR.npz',
                    fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_zcen7.5_noSFR.npz',
                    fileloc_25Mpc+'emission_halpha_L0025N0376_28_SmAb_C2Sm_8000pix_5.000000slice_total_noSFR.npz']
```

```python
print('Reading: '+files_noSF_25Mpc[0]+'...')
data1 = (np.load(files_noSF_25Mpc[0])['arr_0'])[sl]
data1 = get_halpha_SB.imreduce(data1, factor, log=True, method = 'average')
print('Reading" "'+files_SF_25Mpc[0]+'...')
data11 = (np.load(files_SF_25Mpc[0])['arr_0'])[sl]
data11 = get_halpha_SB.imreduce(data11, factor, log=True, method = 'average')
print('5 Mpc slice...')
data_25Mpc = np.log10(10**data1+10**data11)
print('delete %s...'%(files_noSF_25Mpc[0]))
del data11
```

```python
fig = plt.subplots(1,1,figsize=[10,10])
bins = np.arange(-7,7,14./50.)
plt.hist(data2.flatten(),bins=bins,log='True',normed='True',histtype='step',
         label='noSF 100Mpc',color= 'red')
plt.hist(data_28.flatten(),bins=bins,log='True',normed='True',histtype='step',
         label='with SF 100Mpc',color='red',linestyle='--')
plt.hist(data1.flatten(),bins=bins,log='True',normed='True',histtype='step',
         label='noSF 25Mpc',color='blue')
plt.hist(data_25Mpc.flatten(),bins=bins,log='True',normed='True',histtype='step',
         label='with SF 25Mpc',color='blue',linestyle='--')
plt.ylabel(r'$N^{-1}_{pix}dN_{pix}/dlog_{10}S_{B}$')
plt.xlabel(r'$log_{10}S_{B}$ ($ph$ $s^{-1}$ $cm^{-2}$ $sr^{-1})$')
plt.legend()
plt.savefig('plot_SBhist_100Mpc15MpcComparison.pdf')
plt.show()
```

### Conclusion:

Histograms are quite similar to one another.  The shapes are almost identical.  The slight offset could just be due to the different box sizes.  The larger box size might better represent the actual ratio of filaments to voids, for example.

Projections on niagara are fine!
